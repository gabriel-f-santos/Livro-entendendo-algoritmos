# Grafos

Um grafo é um tipo de estrutura de dado onde podemos modelar problemas relacionando objetos entre si. 

Por exemplo: Relação entre rotas para sair de um lugar e chegar a outro, relação de amigos em uma rede social, entre outros...

A figura abaixo apresenta exemplos de grafos, onde os círculos, que representam alguma entidade são chamados de vértice e a relação/conexão entre eles é chamada de aresta

**Pesquisa em Largura** utiliza grafos, é usado para responder dois tipos de perguntas:

- Existe algum caminho do vértice A para o vértice B?

- Qual o caminho mínimo do vértice A até o vértice B?

#Fila

É uma estrutura de dados do tipo FIFO (first in first out), ela simula realmente uma abstração de fila da vida real, onde os primeiros a entrarem na fila são os primeiros a chegarem ao destino final da fila.

Para implementar um grafo, devemos ter um vértice e o mapeamento de todos os seus vértices vizinhos, para isso, deixando cada vértice único, podemos usar uma tabela hash (dicionário), onde nas chaves temos o vértice e nos valores uma lista de vértices vizinhos (arestas que apontam do vértice atual para outro).





```python
def pessoa_eh_vendedor(nome):
  return nome[-1] == 'm'
```

```python
grafo = {}
grafo['gabriel'] = ['claire', 'renam', 'cleiton']
grafo['claire'] = ['jonas', 'renam', 'mauricio']
grafo['cleiton'] = ['julio', 'tyriom', 'michael']
grafo['renam'] = ['julio', 'tyriom']
grafo['jonas'] = ['julio', 'michael']
grafo['mauricio'] = ['tyriom', 'michael']
grafo['julio'] = ['michael']
grafo['tyriom'] = ['julio', 'michael']
grafo['michael'] = ['julio', 'tyriom']
```

```python
from collections import deque
def pesquisa(nome):
  fila_de_pesquisa = deque()
  fila_de_pesquisa += grafo[nome]
  verificadas = []
  while fila_de_pesquisa:
    pessoa = fila_de_pesquisa.popleft()
    if not pessoa in verificadas:
      if pessoa_eh_vendedor(pessoa):
        verificadas.append(pessoa)
        print(f'a pessoa {pessoa} é vendedor de manga / verificadas: {verificadas}')
        # return True
      else:
        fila_de_pesquisa += grafo[pessoa]
        verificadas.append(pessoa)
        print(f'verificadas: {verificadas}')
  # return False
```

```python
pesquisa('gabriel')
```
