## 1.1 - Pesquisa Binária

**Pensamento:** Se estivermos procurando um nome em uma lista telefônica (se você entende essa referência, eu sei, estamos ficando velhos), como a lista é ordenada em ordem alfabética, faz sentido abrir essa lista na metade e decidir para qual metade seguir a busca.

A pesquisa binária é um algoritmo que age de modo semelhante, recebendo uma lista ordenada de valores e retornando a posição do elemento se ele estiver na lista ou None caso contrário.
A pesquisa binária vai sempre selecionar valores do meio de determinar se o valor selecionado está em uma metade ou outro, sempre descartando metade do espaço de busca. 

A pesquisa binária leva log(n) etapas para retornar o resultado, sendo n o tamanho da lista de entrada. Considerando uma lista de 8 valores, log8 = x, assim 2^x = 8, resolvendo, temos que x=3, então serão necessárias 3 iterações até encontrar um resultado **no pior caso**, se fosse uma busca linear (passando por cada elemento da lista, no pior caso teríamos 8 iterações. Esse estudo mostra a importância da complexidade de algoritmos.

Usando a notação big O, que mede quanto o crescimento de tempo de execução de um algoritmo cresce de acordo com a entrada:

Tempo de execução pesquisa simples: linear O(n)

Tempo de execução pesquisa binária: O(log(n))

```python
import math
def pesquisa_binaria(lista, item):
  valor_inferior = 0
  valor_superior = len(lista) - 1
  cont = 0
  while(valor_inferior <= valor_superior):
    valor_central =  int((valor_inferior + valor_superior)/2)
    chute = lista[valor_central]
    if chute == item:
      return valor_central
    if chute < item:
      valor_inferior = valor_central + 1
    else:
      valor_superior = valor_central - 1
  return None

def log(valor, base):
  return math.ceil(math.log(valor, base))

lista = list(range(20))
item = 19
pos = pesquisa_binaria(lista, item)
n_iters = log(len(lista), 2)
print(f'O item {item} foi encontrado na posição {pos} após {n_iters} iterações')
```

```python
def pesquisa_simples(lista, item):
  for i in lista:
    if lista[i] == item:
      return i
  return None
```

```python
import time
lista = list(range(100000000))
item = -1
```

```python
start_time = time.time()
pesquisa_simples(lista, item)
print(f"pesquisa simples - {(time.time() - start_time)} segundos")
```

```python
start_time = time.time()
pesquisa_binaria(lista, item)
print(f"pesquisa binaria - {(time.time() - start_time)} segundos")
```
